### Viết chương trình giải phương trình bậc 2 (ax^2+bx+c=0)

```python
import math


a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))

if a == 0:
    
    if b == 0:
        print("Vô số nghiệm" if c == 0 else "Vô nghiệm")
    else:
        print("Phương trình bậc nhất có nghiệm;" -c/b)
else:

    delta = b**2 - 4*a*c
    print("delta:", delta)

    if delta < 0:
        print("Phương trình vô nghiệm thực")
    elif delta == 0:
        
        print("Phương trình có nghiệm kép:", -b / (2 * a))
    else:
       
        print(f"Phương trình có 2 nghiệm phân biệt:")
        print("x1 =", (-b + math.sqrt(delta)) / (2 * a))
        print("x2 =", (-b - math.sqrt(delta)) / (2 * a))
```

```python
import math

# Nhập các hệ số
a,b,c = map(float,input().split())


if a == 0:
    # Nếu a = 0, phương trình trở thành bx + c = 0
    if b == 0:
        print("Vô số nghiệm" if c == 0 else "Vô nghiệm")
    else:
        print(f"Phương trình bậc nhất có nghiệm x = {-c/b}")
else:
    # Tính Delta
    delta = b**2 - 4*a*c
    print(f"Delta = {delta}")

    if delta < 0:
        print("Phương trình vô nghiệm thực")
    elif delta == 0:
        x = -b / (2 * a)
        print(f"Phương trình có nghiệm kép x1 = x2 = {x}")
    else:
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        print(f"Phương trình có 2 nghiệm phân biệt:")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")
```

### Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math

print("--- KIỂM TRA VỊ TRÍ ĐIỂM A(x, y) VỚI ĐƯỜNG TRÒN ---")

x_i = float(input("Nhập hoành độ tâm I (x_i): "))
y_i = float(input("Nhập tung độ tâm I (y_i): "))
r = float(input("Nhập bán kính R: "))

x_a = float(input("Nhập hoành độ điểm A (x): "))
y_a = float(input("Nhập tung độ điểm A (y): "))

d_binh_phuong = (x_a - x_i)**2 + (y_a - y_i)**2
r_binh_phuong = r**2


print("-" * 30)
if d_binh_phuong < r_binh_phuong:
    print("=> Kết quả: Điểm A nằm TRONG đường tròn.")
elif d_binh_phuong == r_binh_phuong:
    print("=> Kết quả: Điểm A nằm TRÊN đường tròn (thuộc đường tròn).")
else:
    print("=> Kết quả: Điểm A nằm NGOÀI đường tròn.")


khoang_cach = math.sqrt(d_binh_phuong)
print(f"(Khoảng cách thực tế IA ≈ {round(khoang_cach, 2)})")
```
