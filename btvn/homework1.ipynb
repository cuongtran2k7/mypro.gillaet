### Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 

```python

a, b = map(float, input("Nhập chiều rộng a và chiều dài b: ").split())

r = a / 2
s_hcn = a * b
s_tron = 3.14 * r * r
s_con_lai = s_hcn - s_tron

print(f"Diện tích còn lại: {s_con_lai:.2f}")
```

###  Từ 0 – 50kWh: 1.984 đồng/kWh 
###  Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh 
###  Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh 
###  Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh 
###  Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh 
###  Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 
### Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng 

```python

ho_ten = input("Nhập họ tên chủ hộ: ")
chi_so_cu = float(input("Nhập chỉ số điện tháng trước: "))
chi_so_moi = float(input("Nhập chỉ số điện tháng này: "))


so_dien = chi_so_moi - chi_so_cu
tien_chua_vat = 0

# 3. Tính tiền theo từng bậc (Lũy tiến)
if so_dien <= 50:
    tien_chua_vat = so_dien * 1984
elif so_dien <= 100:
    tien_chua_vat = 50 * 1984 + (so_dien - 50) * 2050
elif so_dien <= 200:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + (so_dien - 100) * 2380
elif so_dien <= 300:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + 100 * 2380 + (so_dien - 200) * 2998
elif so_dien <= 400:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (so_dien - 300) * 3350
else:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (so_dien - 400) * 3460

thue_vat = tien_chua_vat * 0.08
tong_tien = tien_chua_vat + thue_vat
print("----------------------------")
print("Chủ hộ:", ho_ten)
print("Số điện tiêu thụ:", so_dien, "kWh")
print("Tổng tiền điện (đã bao gồm 8% VAT):", int(tong_tien), "đồng")
```

### Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) 

```python
def chuyen_so_thanh_chu(n):
    chu_so = ["Không", "Một", "Hai", "Ba", "Bốn", "Năm", "Sáu", "Bảy", "Tám", "Chín"]
    
    if 0 <= n <= 9:
        print(f"{n} -> {chu_so[n]}")
    else:
        print("Vui lòng chỉ nhập số từ 0 đến 9!")
n = int(input("Nhập một số nguyên (0-9): "))
chuyen_so_thanh_chu(n)
```

```python
def chuyen_so_thanh_chu():
    try:

        so = int(input("Nhập một số nguyên từ 0 đến 9: "))

        match so:
            case 0:
                print("Không")
            case 1:
                print("Một")
            case 2:
                print("Hai")
            case 3:
                print("Ba")
            case 4:
                print("Bốn")
            case 5:
                print("Năm")
            case 6:
                print("Sáu")
            case 7:
                print("Bảy")
            case 8:
                print("Tám")
            case 9:
                print("Chín")
            case _: 
                print("Số nhập vào không nằm trong khoảng từ 0 đến 9!")
                
    except ValueError:
        print("Lỗi: Vui lòng nhập vào một số nguyên!")


chuyen_so_thanh_chu()
```
