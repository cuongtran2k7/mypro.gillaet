#### W3A4
### Viết chương trình chia một số m cho n, kết quả thu được làm tròn xuống


```python
n = 0.666666
k = 3

x = n * (10 ** (k + 1))  
x = int(x)            

last_digit = x % 10      # 6

if last_digit >= 5:
    result = (x // 10) + 1
else:
    result = x // 10

final_output = result / (10 ** k)

print(f"Số ban đầu: {n}")
print(f"Kết quả sau làm tròn {k} chữ số: {final_output}")
```

#### W3A11
### Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường". 

```python
a,b,c=map(int,input().split())
if a + b > c and a + c > b and b + c > a:

    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:

        if (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
            print("Tam giác vuông cân")
        else:
            print("Tam giác cân")
    elif (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
        print("Tam giác vuông")
    else:
        print("Tam giác thường")

else:
    print("Không phải tam giác")
```

#### W3A13
### Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau: 
### 0-50 kWh: 1500đ/kWh 
### 51-100 kWh: 2000đ/kWh 
### 100 kWh: 3000đ/kWh 


```python

kwh = float(input("Nhập số kWh điện tiêu thụ: "))

tong_tien = 0

if kwh <= 50:
    # Bậc 1
    tong_tien = kwh * 1500
elif kwh <= 100:
    # Bậc 2
    tong_tien = (50 * 1500) + (kwh - 50) * 2000
else:
    # Bậc 3
    tong_tien = (50 * 1500) + (50 * 2000) + (kwh - 100) * 3000
print(f"Tổng tiền điện phải trả: {tong_tien:,.0f}đ")
```

### Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm sẵn có của python). Các kết quả cách nhau một dấu cách 

```python
n = float(input())

phan_nguyen = int(n)
if n == phan_nguyen:
    lam_tron_len = phan_nguyen
    lam_tron_xuong = phan_nguyen
    lam_tron_gan_nhat = phan_nguyen
elif n > 0:
    lam_tron_len = phan_nguyen + 1
    lam_tron_xuong = phan_nguyen
    lam_tron_gan_nhat = int(n + 0.5)

else:
    lam_tron_len = phan_nguyen
    lam_tron_xuong = phan_nguyen - 1
    lam_tron_gan_nhat = int(n - 0.5)
print(f"{lam_tron_len} {lam_tron_xuong} {lam_tron_gan_nhat}")
```

#### W3A17
### Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không? 


```python
a,b,c,d=map(int,input().split())

# th a=0
if a == 0:
    if b == 0 and c == 0 and d == 0:
        print("YES - Là cấp số nhân (công bội tùy ý)")
    else:
        print("NO - Không phải cấp số nhân")
else:
    # Kiểm tra b có chia hết cho a không
    if b % a == 0:
        # Tìm công bội nguyên
        cong_boi = b // a 
        if b * cong_boi == c and c * cong_boi == d:
            print(f"YES - Là cấp số nhân với công bội nguyên q = {cong_boi}")
        else:
            print("NO - Không phải cấp số nhân")
    else:
        print("NO - Công bội không phải là số nguyên")
```
